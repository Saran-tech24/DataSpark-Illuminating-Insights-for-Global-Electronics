```python
mycursor.execute("use project2")
mycursor.execute("""
    SELECT stores.Country,YEAR(stores.Open_Date) AS year,DATE_FORMAT(stores.Open_Date, '%M') AS MONTHS,COUNT(stores.Store_Key) AS stores_counts,
        AVG(stores.Square_Meters),SUM(sales.Quantity * products.Unit_Price_USD) AS TotalRevenue
    FROM  stores
    JOIN  sales ON stores.Store_Key = sales.Store_key
    JOIN  products ON sales.Product_key = products.Product_key
    GROUP BY year, MONTHS, stores.Country
    order by TotalRevenue asc""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```
